### 3. What are top 3 apps that have highest number of users installed/download among all categories?

```python
data = gp[['Category','App','Installs','Rating','Reviews']]
data2 = data[data['Installs'] != 'Free'] #Remove 'Free'
data2.drop_duplicates(subset='App', inplace=True) #Drop duplicate based on 'App'
data2['Installs'] = data2['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x) #Remove '+'
data2['Installs'] = data2['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x) #Remove '-'
data2['Installs'] = data2['Installs'].astype(int) #Change data type
sorting = data2.sort_values('Installs', ascending=False)
sorting.head(3)
```

### 4. What are top 3 apps that have highest number of users installed/download among the following categories: (Hint: use installs column)

```python
cat = ['BUSINESS','COMMUNICATION','GAME','DATING','MUSIC']
new = data2[data2['Category'].isin(cat)]
selected_data = new[['Category','App','Installs']]
sorted_data = selected_data.sort_values(by=['Category','Installs'],ascending=False)
sorted_data.groupby('Category').head(3)
```

### 5. What are top 3 apps that have highest rating among the following categories:

```python
selected_data1 = new[['Category','App','Installs','Rating']]
sorted_data1 = selected_data1.sort_values(by=['Category','Rating'], ascending=False)
sorted_data1.groupby('Category').head(3)
```
