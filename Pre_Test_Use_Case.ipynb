```python
#Upload Data
import pandas as pd
telco = pd.read_csv('churn.csv', encoding='latin1')
telco.head(5)
```

```python
# 1. Bagaimanakan rata- rata pemakaian pelanggan kategori tidak churn ('class=0')?
telco.loc[telco['class']==0,['total_day_minutes','total_day_calls','total_day_charge','total_eve_minutes',\
             'total_eve_calls','total_eve_charge','total_night_minutes','total_night_calls','total_night_charge',\
             'total_intl_minutes','total_intl_calls','total_intl_charge','number_customer_service_calls']].mean()
```

```python
# 2. Bagaimanakan rata- rata pemakaian pelanggan berdasarkan kategori churn ('class'=1)?
telco.loc[telco['class']==1,['total_day_minutes','total_day_calls','total_day_charge','total_eve_minutes',\
             'total_eve_calls','total_eve_charge','total_night_minutes','total_night_calls','total_night_charge',\
             'total_intl_minutes','total_intl_calls','total_intl_charge','number_customer_service_calls']].mean()
```

```python
# 3. Sebutkan variabel yang memiliki perbedaan signifikan dalam atas pelanggan churn dan stay?
from scipy import stats
numeric_variable = ['total_day_minutes','total_day_calls','total_day_charge','total_eve_minutes',\
             'total_eve_calls','total_eve_charge','total_night_minutes','total_night_calls','total_night_charge',\
             'total_intl_minutes','total_intl_calls','total_intl_charge','number_customer_service_calls']

churn=telco[telco['class']==1]
stay=telco[telco['class']==0]
from scipy import stats

for item in numeric_variable:
    a = stats.ttest_ind(churn[item], stay[item],  equal_var = False)
    if (a.pvalue<0.05):
        print(item)
```

```python
# 4. Apakah variabel - variabel pada poin 3 saling berkorelasi?
telco.loc[:,numeric_variable].corr(method= 'spearman')
```
